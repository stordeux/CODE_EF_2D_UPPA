# Les matrices sur un élément


## La matrice de masse locale
$$
M_{i,j} = \int_{T} w_i(x,y)\,w_j(x,y)\,dx\,dy
= \left| \det\big(A_2-A_1,\;A_3-A_1\big) \right|\,\widehat{M}_{i,j}
$$
avec $T$ un triangle de sommets $A_1$, $A_2$, $A_3$.

## La matrice mixte de reference
$$
K^{(T,x)}_{i,j}=\int_{T}
w_{m_1,n_1}(x,y)\,
\frac{\partial w_{m_2,n_2}}{\partial x}(x,y)\,dx\,dy
$$
$$
K^{(T,y)}_{i,j}=\int_{T}
w_{m_1,n_1}(x,y)\,
\frac{\partial w_{m_2,n_2}}{\partial y}(x,y)\,dx\,dy
$$
avec
$$
0\le i < N_{\text{loc}},\quad 
0\le j < N_{\text{loc}},\quad
i\sim(m_1,n_1)\ \text{et}\ j\sim(m_2,n_2).
$$

## La matrice de rigidite locale
$$
K^{(T,xx)}_{i,j}=\int_T
\frac{\partial w_{m_1,n_1}}{\partial x}(x,y)\,
\frac{\partial w_{m_2,n_2}}{\partial x}(x,y)\,dx\,dy
$$
$$
K^{(T,xy)}_{i,j}=\int_T
\frac{\partial w_{m_1,n_1}}{\partial x}(x,y)\,
\frac{\partial w_{m_2,n_2}}{\partial y}(x,y)\,dx\,dy
$$
$$
K^{(T,yx)}_{i,j}=\int_T
\frac{\partial w_{m_1,n_1}}{\partial y}(x,y)\,
\frac{\partial w_{m_2,n_2}}{\partial x}(x,y)\,dx\,dy
$$
$$
K^{(T,yy)}_{i,j}=\int_T
\frac{\partial w_{m_1,n_1}}{\partial y}(x,y)\,
\frac{\partial w_{m_2,n_2}}{\partial y}(x,y)\,dx\,dy
$$
