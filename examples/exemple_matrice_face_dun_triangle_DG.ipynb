# Matrice locale de face

Matrice de masse de frontière sur une face $F$ de l’élément $T$ (DG). Pour les fonctions de base locales $\varphi_i^{(T)}$ de $T$, on définit :

$$
M_F^{(T)} = \bigl[(M_F^{(T)})_{ij}\bigr]_{i,j=1}^{N_{\text{loc}}},
\quad (M_F^{(T)})_{ij} = \int_F \varphi_i^{(T)}\,\varphi_j^{(T)}\,ds.
$$

Cette matrice mesure la contribution de la face $F$ dans les intégrales de bord (termes de flux, pénalisation, etc.).
