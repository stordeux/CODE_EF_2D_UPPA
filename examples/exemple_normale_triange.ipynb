# Construction des normales à un triangle

On considère un triangle $A_0$, $A_1$, $A_2$.
Celui-ci admet trois côtés dont les normales sortantes sont 

Face 0 : $n_{0}\perp [A_1,A_2]$

Face 1 : $n_{1}\perp [A_2,A_0]$

Face 2 : $n_{2}\perp [A_0,A_1]$


```python
# Test avec un triangle quelconque
A0 = np.array([0.0, 0.0],dtype=float)
A1 = np.array([1.0, 0.0],dtype=float)
A2 = np.array([0.5, 0.8],dtype=float)

print("Test de la fonction calcul_normale\n")
print(f"Triangle : A0={A0}, A1={A1}, A2={A2}\n")

for i in range(3):
    n = calcul_normale(A0, A1, A2, i)
    print(f"Normale opposée à A{i} : n{i} = ({n[0]:.6f}, {n[1]:.6f})")
    print(f"  Norme : {np.linalg.norm(n):.6f}")

plot_triangle_with_normals(A0, A1, A2)
```
