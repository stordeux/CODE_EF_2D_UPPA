# Intégrale sur un segment qui relie les points $A_1$ et $A_2$ par méthode de Gauss

La quadrature de Gauss sur $[-1,1]$ approxime l'integrale par une somme ponderee :
$$
\int_{-1}^{1} f(\xi)\,d\xi \approx \sum_{i=1}^{n} w_i\,f(\xi_i),
$$
ou $\xi_i$ sont les racines du polynome de Legendre d'ordre $n$ et $w_i$ les poids associes. Cette formule est exacte pour tout polynome de degre $\le 2n-1$.

Pour un segment 2D reliant $A_1=(x_1,y_1)$ a $A_2=(x_2,y_2)$, on le parametre par
$$
\mathbf{r}(\xi)=\frac{1-\xi}{2}A_1+\frac{1+\xi}{2}A_2,\quad \xi\in[-1,1],
$$
avec la longueur $L=\|A_2-A_1\|$. L'integrale le long du segment s'ecrit
$$
\int_{A_1}^{A_2} f(x,y)\,ds = \frac{L}{2}\int_{-1}^{1} f\big(\mathbf{r}(\xi)\big)\,d\xi
\approx \frac{L}{2}\sum_{i=1}^{n} w_i\, f\big(\mathbf{r}(\xi_i)\big),
$$
ou $(\xi_i,w_i)$ sont les points et poids de Gauss sur $[-1,1]$.

```python
# Exemple d'utilisation
A1 = (1, 1)
A2 = (1, 2)

def f(x, y, m, n):
    return x**m * y**n

n=7
xi, w = leggauss(n)

Iold = integrate_segment_2D_old(lambda x, y: f(x, y, m=2, n=3), A1, A2, xi, w)
I = integrate_segment_2D(lambda x, y: f(x, y, m=2, n=3), A1, A2, xi, w)
TEST = np.allclose(I, Iold)
print(f"Intégrale de f(x,y)=x^2y^3 sur le segment de {A1} à {A2} : {I}")
print("Comparaison avec l'ancienne méthode :", TEST)
```

**Quadrature de Gauss 2D sur un triangle par integrate_triangle_2D**

On approxime l’integrale par une double somme ponderee :
$$
\int_T f(x,y) dxdy \approx \sum_{i=1}^{n}\sum_{j=1}^{n} w_i\,w_j\, f(x_{ij},y_{ij})\, J_{ij}
$$

**Definitions :**
- $x_{ij},y_{ij}$ sont les coordonnees du point du triangle obtenu par changement de variables.
- $J_{ij}$ est le Jacobien de la transformation, qui inclut le facteur $(1-s)$.
- $w_i, w_j$ sont les poids de Gauss sur $[-1,1]$.

**Changement de variables :**
$$
s = \frac{1+\xi_i}{2}, \quad t = \frac{1+\xi_j}{2}
$$
$$
(x_{ij},y_{ij}) = A_1 + s(A_2-A_1) + t(1-s)(A_3-A_1)
$$
$$
J_{ij} = \text{aire}(A_1A_2A_3)\,\frac{1-s}{2}
$$

```python
def f(x, y, m=1, n=1):
    return x**m * y**n
print(f(3,2))
A1 = (0, 0)
A2 = (1, 0)
A3 = (0, 1)
I = integrate_triangle_2D(lambda x, y: f(x, y, m=1, n=2), A1, A2, A3, xi, w)
print(f"Intégrale de f(x,y)=1 sur le triangle de sommets {A1}, {A2}, {A3} : {I}")


```
